# Fill Nan in PropertyAddress Column 

**Check is there any Nan value**

```python
Nashville_housing["PropertyAddress"].isna().any()
```

```python
Nashville_housing.sort_values(by = "ParcelID").head(1000)
```

**Get ParcelID of Nan PropertyAddress column**

```python
nan_parcelid = Nashville_housing.loc[Nashville_housing["PropertyAddress"].isna(), ["ParcelID"]]
```

```python
nan_parcelid
```

**Get the repeated PacelID that it PropertyAddress is Nan**

```python
Nashville_housing["fill_nan"] = Nashville_housing.sort_values("ParcelID").groupby("ParcelID").cumcount() + 1
```

```python
repeated_parcelid = Nashville_housing.loc[(Nashville_housing["fill_nan"] > 1) & (Nashville_housing["PropertyAddress"].isna()), ["ParcelID"]]
```

```python
repeated_parcelid
```

**Merge a and Nashville_housing["fill_na"] to check the match**

```python
pd.merge(nan_parcelid, repeated_parcelid, on = "ParcelID", how = "inner")
```

**Everything is match**

**Which mean every repeated ParcelID will have one PropertyAddress that is Nan**

```python
Nashville_housing["PropertyAddress"].ffill(inplace = True)
```
