## Data preperation Lebron James stat

```python
x = pd.read_html(r"https://www.basketball-reference.com/players/j/jamesle01.html")
```

```python
x[4]
```

**Merge x[1] and x[4]**

```python
lebron_stat = pd.merge(x[1], x[4], left_index = True, right_index = True, how = "inner")[["Season_x", "Age_x", "Tm_x", "PTS_x", "PTS_y"]]
```

```python
lebron_stat
```

**Rename columns**

```python
lebron_stat.rename(columns = {"Season_x": "Season", "Age_x": "Age", "Tm_x": "Team", "PTS_x": "PPG", "PTS_y": "PTS"}, inplace = True)
```

```python
lebron_stat
```

**Drop unused rows**

```python
lebron_stat.drop(lebron_stat.index[22:], inplace = True)
```

```python
lebron_stat
```

**Sum all PTS Lebron James made**

```python
lebron_stat["Career PTS"] = lebron_stat["PTS"].cumsum()
```

```python
lebron_stat
```

```python
lebron_stat.loc[[21], ["Career PTS"]] = lebron_stat.loc[21, "PTS"]
```

# LeBron James VS Active elite scorer

---

## Data preperation Active elite scorer

```python
# NBA all time PPG record 

z = pd.read_html(r"https://www.basketball-reference.com/leaders/pts_per_g_career.html")
```

```python
len(z)
```

```python
all_time_ppg_rec = z[2]

all_time_ppg_rec
```

__Replace * with nothing__

```python
all_time_ppg_rec["Player"] = all_time_ppg_rec["Player"].str.replace("*", "")
```

### all_time_ppg_rec

```python
all_time_ppg_rec
```

```python
# Current active NBA player

a = pd.read_html(r"https://www.basketball-reference.com/leagues/NBA_2024_per_game.html")
```

```python
a[0]
```

```python
current_active_player = a[0]
```

### current_active_player

```python
current_active_player
```

---

## Find top 10 current active NBA player with highest career points

**Merge current_active_player and all_time_scoring_rec and all_time_ppg_rec**

```python
b = pd.merge(all_time_scoring_rec, current_active_player, on = "Player", how = "inner")[["Rank", "Player", "Age","PTS_x"]].rename(columns = {"PTS_x": "Career PTS"}).head(10)

b
```

```python
top_10 = pd.merge(b, all_time_ppg_rec, on = "Player", how = "inner")[["Rank_x", "Player", "Age", "Career PTS", "PPG"]].rename(columns = {"Rank_x": "Rank"})

top_10
```

```python
top_10.dtypes
```

**Change Age to int32**

```python
top_10 = top_10.astype({"Age": "int32"})
```

### top_10

```python
top_10
```

---

# Find how many season these player need to tide current Lebron James record 


**Create new column name season_to_tide**

```python
top_10["season_to_tide"] = 0

top_10
```

**Assume condition**
- **These player only miss 10 games in a season**
- **They keep scoring PPG as same as they are now**

```python
top_10.loc[1:, "season_to_tide"] = ((top_10.loc[0, "Career PTS"] - top_10["Career PTS"]) / (top_10["PPG"] * 72)).round()

top_10
```

**Create new column age_to_tide to find their age when they tied current Lebron James scoring record**

```python
top_10["age_to_tide"] = top_10["Age"] + top_10["season_to_tide"]
```

**Merge top_10 and lebron_stat**
- merge on Age to compare ability to score at that age of each player to LeBron James when he is same age as them

```python
top_10 = pd.merge(top_10, lebron_stat[["Age", "Career PTS"]], on = "Age", how = "inner").rename(columns = {"Career PTS_y": "LeBron_PTS_at_player_age","Career PTS_x": "Career PTS", "PPG": "Career PPG"})
```

### Top_10

```python
top_10
```

## From the table above it nearly impossible or even impossible for anybody in the table to break LeBron James scoring record

**Create xpos and xticks for graph ploting**

```python
xpos = np.arange(len(top_10.loc[1:, "Player"])) 
xticks = [x for x in top_10.loc[1:, "Player"]]
```

```python
xpos
```

```python
xticks
```

```python
plt.style.use("dark_background")

font_label = {"family": "Courier New",
                    "color": "white",
                    "weight": "bold",
                    "size": "15"}

font_text = {"family": "Courier New",
                   "color": "orangered",
                   "weight": "bold",
                   "size": "12"}

font_title = {"family": "Courier New",
                    "color": "white",
                    "weight": "bold",
                    "size": "20"}

plt.figure(figsize = (15,10))

plt.bar(xpos, top_10.loc[1:, "LeBron_PTS_at_player_age"], width = 0.3, label = "LeBron James")
plt.bar(xpos +0.3, top_10.loc[1:, "Career PTS"], width = 0.3, label = "Player")


plt.xticks(xpos, xticks, rotation = 45)
plt.yticks(np.arange(0, 45001, 5000))

plt.xlabel("Player", fontdict = font_label)
plt.ylabel("PTS", fontdict = font_label)
plt.title("LeBron James VS Active elite scorer(at their age)", fontdict = font_title)

for i in xpos:
    plt.text(i, top_10.loc[i + 1, "LeBron_PTS_at_player_age"], f'{top_10.loc[i + 1, "Age"]} years old', fontdict = font_text, ha = "center", va = "bottom", rotation = 30)

plt.grid(False)
    
plt.legend(fontsize = 15)
plt.show()
```

## From the graph above LeBron James win against every player in scoring at their age

# LeBron James VS Young talent player

---

![young talent](https://hoopshype.com/wp-content/uploads/sites/92/2022/01/Best_Under25b.png?w=1000&h=600&crop=1)

```python
# Top 100 current active player that score most point

top_100_current = pd.read_html(r"https://www.basketball-reference.com/leaders/pts_active.html")[0]

top_100_current
```

**Merge top_100_current and current_active_player**

```python
young_potential = pd.merge(top_100_current, current_active_player[["Age", "Player"]], on = "Player").rename(columns = {"PTS": "Career PTS"})

young_potential
```

**Change Age to int32**

```python
young_potential = young_potential.astype({"Age": "int32"})
```

### At the age of 25 Lebron James already score 15251 PTS  

**Find**
- Player whose age is less than or equal to 27
    - AND 
- Score just about 50% of LeBron James PTS at his age of 25

```python
young_potential = young_potential.loc[(young_potential["Age"] <= 27) & (young_potential["Career PTS"] > 15251 / 2)].reset_index(drop = True)

young_potential.sort_values(by = "Career PTS", ascending = False)
```

**Import data of each player career PPG**

```python
DB = pd.read_html(r"https://www.basketball-reference.com/players/b/bookede01.html")[1]
DB["Player"] = "Devin Booker"
DB = DB.loc[[9], ["PTS", "Player"]]

DM = pd.read_html(r"https://www.basketball-reference.com/players/m/mitchdo01.html")[1]
DM["Player"] = "Donovan Mitchell"
DM = DM.loc[[7], ["PTS", "Player"]]

JT = pd.read_html(r"https://www.basketball-reference.com/players/t/tatumja01.html")[1]
JT["Player"] = "Jayson Tatum"
JT = JT.loc[[7], ["PTS", "Player"]]

LD = pd.read_html(r"https://www.basketball-reference.com/players/d/doncilu01.html")[1]
LD["Player"] = "Luka Dončić"
LD = LD.loc[[6], ["PTS", "Player"]]

TY = pd.read_html(r"https://www.basketball-reference.com/players/y/youngtr01.html")[1]
TY["Player"] = "Trae Young"
TY = TY.loc[[6], ["PTS", "Player"]]

DR = pd.read_html(r"https://www.basketball-reference.com/players/r/russeda01.html")[1]
DR["Player"] = "D'Angelo Russell"
DR = DR.loc[[13], ["PTS", "Player"]]

JB = pd.read_html(r"https://www.basketball-reference.com/players/b/brownja02.html")[1]
JB["Player"] = "Jaylen Brown"
JB = JB.loc[[8], ["PTS", "Player"]]

DF = pd.read_html(r"https://www.basketball-reference.com/players/f/foxde01.html")[1]
DF["Player"] = "De'Aaron Fox"
DF = DF.loc[[7], ["PTS", "Player"]]

BI = pd.read_html(r"https://www.basketball-reference.com/players/i/ingrabr01.html")[1]
BI["Player"] = "Brandon Ingram"
BI = BI.loc[[8], ["PTS", "Player"]]

DS = pd.read_html(r"https://www.basketball-reference.com/players/s/sabondo01.html")[1]
DS["Player"] = "Domantas Sabonis"
DS = DS.loc[[10], ["PTS", "Player"]]
```

```python
young_player = pd.concat([DB, DM, JT, LD, TY, DR, JB, DF, BI, DS])

young_player
```

**Merge young_potential and young_player**

```python
young_talent = pd.merge(young_potential, young_player, on = "Player", how = "inner").rename(columns = ({"PTS": "Career PPG"}))[["Rank", "Player", "Career PTS", "Career PPG", "Age"]]
```
